## Уровень 0

посчитайте значение 1-ой и 2-ой производной фунции

f(x) = x^5 + 4sin(2x) + cos(3x+3) 

в точке x = 1


```python
from sympy import *
```

```python
x = Symbol('x')
y = (x**5 + 4*sin(2*x) + cos(3*x+3))
y_first = y.diff(x)
y_second = y_first.diff(x)
print('Первая производная функции: ', y_first)
print('Вторая производная функции: ', y_second)
```

```python
def f1(x):
    return 5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x)
def f2(x):
    return 20*x**3 - 16*sin(2*x) - 9*cos(3*x + 3)
print(f'При х = 1, значение первой производной: {f1(1)}')
print(f'Значение второй производной: {f2(1)}')
```

## Уровень 1

Постройте график первой и второй производной функции

(sin(2x+1))^5

на отрезке [-5; 5]

```python
x = Symbol('x')
y = ((sin(2*x+1))**5)
y_first = y.diff(x)
y_second = y_first.diff(x)
print('Первая производная функции: ', y_first)
print('Вторая производная функции: ', y_second)
```

```python
def f1(x):
    return 10*sin(2*x + 1)**4*cos(2*x + 1)
def f2(x):
    return -20*sin(2*x + 1)**5 + 80*sin(2*x + 1)**3*cos(2*x + 1)**2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-5, 5, 100)
y1 = [f1(x1[i]) for i in range(len(x1))]
y2 = [f2(x1[i]) for i in range(len(x1))]  

plt.plot(x1, y1, label= '1-ая производная')
plt.plot(x1, y2, label= '2-ая производная')
plt.title('Графики производной функции')
plt.legend()
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```
